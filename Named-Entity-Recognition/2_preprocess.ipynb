```python
import os
import pandas as pd
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from huggingface_hub import login
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df
```

```python
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)  # 70% train
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)  # 15% val, 15% test
```

```python
train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
valid_ds = Dataset.from_pandas(valid_df.reset_index(drop=True))
test_ds = Dataset.from_pandas(test_df.reset_index(drop=True))
```

```python
dataset = DatasetDict({
    "train": train_ds,
    "validation": valid_ds,
    "test": test_ds
})
```

```python
dataset
```

```python
login()
dataset.push_to_hub("MVesalA/Persian-NER-GUI-Classification")
```
